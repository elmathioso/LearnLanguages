```python
link = 'https://www.manythings.org/vocabulary/lists/a/'

```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'
r = requests.get(link, headers={'User-Agent': user_agent}) 
  
categories = []

soup = BeautifulSoup(r.content, 'lxml') 
data = soup.find_all('li')
for i in data:
    name = i.find('a').text
    newlink = link + i.find('a')['href']
    categories.append({"name":name, "link":newlink})
    
print(categories)
```

```python
pd.DataFrame(categories).to_csv('categories.csv')
```

```python

blocks = 0
alldata = []
for i in categories:
    print("Block:", blocks)
    blocks += 1
    print(i['name'])
    link = i['link']
    print(link)
    r = requests.get(link, headers={'User-Agent': user_agent}) 
    soup = BeautifulSoup(r.content, 'lxml') 
    data = soup.find_all('ul')
    sets = 0
    for j in data:
        print(sets)
        sets += 1
        words = j.find_all('li')
        for k in words:
            print(k.text)
            alldata.append({"block":blocks, "set":sets, "word":k.text, "category":i['name']})
```

```python
pd.DataFrame(alldata)
```

```python
pd.DataFrame(alldata).to_csv('simple_words.tsv', sep='\t')
```

# Scrape Master Russian.com

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
addresses = ['http://masterrussian.com/vocabulary/most_common_words.htm']
addresses += ['http://masterrussian.com/vocabulary/most_common_words_' + str(x) + '.htm' for x in range(2,13)]
addresses
```

```python
user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'


first = True
results = []

for a in addresses:
    r = requests.get(a, headers={'User-Agent': user_agent}) 

    categories = []

    soup = BeautifulSoup(r.content, 'lxml') 
    data = soup.find_all("table", {"class": "topwords"})
    for d in data:
        rows = d.find_all('tr')
        for r in rows:
            t = r.find_all('td')
            if first == True:
                results.append({"number":t[0].text, "word":t[2].text, "translation":t[3].text, "function":t[4].text})
            else:
                results.append({"number":t[0].text, "word":t[1].text, "translation":t[2].text, "function":t[3].text})
                
        first = False
        
print(results)
    #break
```

```python
df = pd.DataFrame(results)
df.to_csv('masterrussian.csv', index=False, encoding='utf8')
```

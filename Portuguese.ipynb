```python
import spacy
import pandas as pd
```

```python
nlp = spacy.load('pt_core_news_lg')
```

```python
df = pd.read_csv('CommonPortugueseWords.csv')
```

```python
df['type'] = ''
for i in df.iterrows():
    print(i)
    word =  nlp(i[1]['pt'])
    if len(word.text.split(' ')) == 1:
        for token in word:
            print(token.pos_)
            df.loc[i[0], 'type'] = token.pos_
```
